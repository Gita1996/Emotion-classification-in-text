```python
import pandas as pd
import re
df = pd.read_csv('ISEAR_dataset2.csv')
pattern=';|:|\*|\d+|#|=|(|)|&'
pattern2='\?|!|\.'
a=df['text']
ll=list()
t1=''
for t in a:
  if t:
    t=re.sub(pattern,'',t)
    t=re.sub(r" +"," ",t)
    sentences=re.findall(pattern2, t)
    for s in sentences:
        for m in s:
            t=t.replace(m,m+"\n")
    lines = t.split("\n")

    non_empty_lines = [line for line in lines if line.strip() != ""]
    print(non_empty_lines)

    t1= ""
    for line in non_empty_lines:

          t1+= line + "\n"
#    t2= [line for line in lines if line.strip() != ""]
  ll.append(t1)
ll2=list()
for n in ll:
  l=''
  for m in n.split():
    if '@' in m:
      m=''
    l+=m+' '
  ll2.append(l)


d={'labels':df['emotion'], 'text':ll2}
df = pd.DataFrame(d)
df.to_csv('ISEAR_dataset2_new.csv', index = False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
df = pd.read_csv('ISEAR_dataset2.csv')
x_train,x_test,y_train,y_test = train_test_split(df["text"], df["labels"],test_size = 0.16)
x_train_l=list()
y_train_l=list()

x_test_l=list()
y_test_l=list()

for x in x_train:
  x_train_l.append(x)

for y in y_train:
  y_train_l.append(y)

for x in x_test:
  x_test_l.append(x)

for y in y_test:
  y_test_l.append(y)

#print(type(x_train))

train_dataset={'labels':y_train_l, 'text':x_train_l}
test_dataset={'labels':y_test_l, 'text':x_test_l}

df = pd.DataFrame(train_dataset)
df.to_csv('text_emotion_train.csv', index = False)

df = pd.DataFrame(test_dataset)
df.to_csv('text_emotion_test1.csv', index = False)



#train_data = pd.concat([x_train , y_train], axis = 1)
```

```python
import json
import pandas as pd

# Opening JSON file
f = open('stop-words-english.json')

# returns JSON object as
# a dictionary
data = json.load(f)
df=pd.read_csv("ISEAR_dataset.csv")
ll=list()
for t in df['text']:
  r=""
  for w in t.split():
    for a in data:
      if w==a:
        w=""
    r+=w+" "
  ll.append(r)

dict2={'emotion':df['emotion'], 'text':ll}
df = pd.DataFrame(dict2)
df.to_csv('ISEAR_dataset_new.csv', index=False)
```
